```python
data.columns

#drop unwanted columns
#data.drop(['TIDAL Popularity','ISRC'], axis=1)
columns_to_delete = [4,5,19,20,21,22,23,24,25,27]
data = data.drop(data.columns[columns_to_delete],axis = 1)
```

```python
#Transform Release Date to datetime format
from datetime import datetime
data['Release Date'] = data['Release Date'].apply(lambda x: datetime.strptime(x,'%m/%d/%Y'))
```

```python
data['Release Date']
```

```python
def from_str_to_int(s):
    if type(s) != type(""):  # Check if the type of s is not a string
        return s
    if '.' in s:  # Check if there's a decimal point in the string
        s = s.replace('.', '')  # Remove all decimal points
    if ',' in s:  # Check if there's a comma in the string
        s = s.replace(',', '')  # Remove all commas
    return int(s)  # Convert the cleaned string to an integer
```

```python
cols_to_int = ['Spotify Streams','Spotify Playlist Count','Spotify Playlist Reach','YouTube Views','YouTube Likes',
               'TikTok Posts','TikTok Likes','TikTok Views','YouTube Playlist Reach','AirPlay Spins','Shazam Counts']
```

```python
for i in cols_to_int:
    data[i] = data[i].apply(from_str_to_int)
```

```python
#Filling in the blanks
data.isnull().sum()
```

```python
#NaN in column Artist we will replace to unknown so its logical
data['Artist'] = data['Artist'].fillna('unknown')
```

```python
cols_obj = ['Track','Album Name','Artist']
for i in data.columns:
    if i not in cols_obj:
        data[i] = data[i].fillna(data[i].median())
```
