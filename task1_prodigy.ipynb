```python
# Drop rows where 'recovered' is null
df = df.dropna(subset=['Recovered'])
# Drop rows where 'deaths' is null
df = df.dropna(subset=['Deaths'])
# Drop rows where 'tests' is null
df = df.dropna(subset=['Tests'])
```

```python
import matplotlib.pyplot as plt
df_sorted = df.sort_values(by='Cases', ascending=True)
df_sorted1 = df_sorted.head(5)
 
# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Plot bar graph
plt.bar(df_sorted1['country'], df_sorted1['Cases'], color='skyblue')

# Add labels and title
plt.xlabel('Country')
plt.ylabel('Cases')
plt.title('Number of COVID-19 Cases by Country')

# Rotate x-axis labels for better readability if necessary
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```
